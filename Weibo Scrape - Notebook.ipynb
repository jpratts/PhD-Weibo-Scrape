```python
# Load packages

from weibo_scraper import  get_weibo_tweets
import pandas as pd
import csv
```

```python
## Create a function which is going to take in a tweet id and the number of pages you want to scrape,
## and then returns a list of tweets.

def scrape_tweets(tweet_id, pages):
    tweets_list = []
    for tweet in get_weibo_tweets(tweet_container_id=str(tweet_id),pages=int(pages)):
        tweets_list.append(tweet)
    return(tweets_list)
```

```python
## Create a function which is going to filter the results

def scrape_clean(name_of_original_output):
    tweets_cleaned = []
    for user in name_of_original_output:
        tweets_cleaned.append(user["mblog"])
    tweets_cleaned = pd.DataFrame(tweets_cleaned)
    tweets_cleaned = tweets_cleaned.dropna(subset=["text"])
    return(tweets_cleaned)
```

```python
## Example of how to use first function.
## 1. Create a variable and pass in the name of the function. 
## 2. in the brackets insert a tweet id, e.g. "10760432034134198", comma, and the number of pages to scrape "1"

tweets_example = scrape_tweets(1076033637346297, 1)
```

```python
## Lets use the next function.
## 1. Create a variable (or reuse old one) and pass in the scrape_clean function. In the brackets pass in the name you used
## in variable 1. 

tweets_cleaned = scrape_clean(tweets_example)
```
